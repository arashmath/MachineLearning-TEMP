```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
# To convert categorical labels to numerical ones using "one hot" encoding
from keras.utils import to_categorical
from keras.backend import argmax
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder as LBE
from sklearn.metrics import confusion_matrix as cm
```

```python
number_of_neurons = [50, 100]
learning_rate = [0.07, 0.5]
batch_size = [64, 128]
repeat = 30
```

```python
import scipy.io
data = scipy.io.loadmat('binaryalphadigs')
print("Primary data type: ", type(data))
print("Keys in dict: ", data.keys(), "\n")

table = data['dat']
print("Main data shape: ", table.shape)
labels = data['classlabels']
print("Shape of labels: ", labels.shape, "\n")

label_array = [i[0][0] for i in labels.T]
print("Labels: ", label_array)
num_classes = len(label_array)
print("Number of classes (The same as number of labels): ", num_classes, "\n")

print("Dimension of a picture: ", table[0][0].shape)
input_dimension = table[0][0].shape[0] * table[0][0].shape[1]
print("Expected size of input array by flattening a picture array: ", input_dimension)
```

```python
# Converting each picture array to a 1-dimension array:
table_to_1d = [[np.matrix.flatten(table[i][j]) for j in range(len(table[i]))] for i in range(len(table))]
print("Picture arrays' dimensions after flattening: ", table_to_1d[0][0].shape, "\n")

# Making a 1-D array of all data, instead of a 36*39 matrix:
table_to_1row = []
for i in range(len(table_to_1d)):
    for j in range(len(table_to_1d[i])):
        table_to_1row.append(table_to_1d[i][j])
X = np.array(table_to_1row)
print("Main data shape after converting it to a 1-row array: ", X.shape, "\n")

# Making an appropriate label array:
temp_labels = np.array(label_array).T
Y = []
for i in range(36):
    for j in range(39):
        Y.append(temp_labels[i])
Y = np.array(Y)
Y = LBE().fit_transform(Y)  # Convert to numerical array
Y = to_categorical(Y, num_classes)  # Encode to one-hot
print("Final labels shape: ",Y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y)
```

```python
# One-layered NN:

all_models1 = []

for n in number_of_neurons:
    for lr in learning_rate:
        for bt_sz in batch_size:
            model = Sequential()
            model.add(Dense(n, input_shape=(input_dimension,), activation='sigmoid'))
            model.add(Dense(num_classes))
    #       model.summary()
            model.compile(optimizer=SGD(lr=lr), loss='mse', metrics=['accuracy'])
            model.fit(x_train, y_train, batch_size=bt_sz, epochs=repeat)

            y_train_pred = [np.argmax(i) for i in model.predict(x_train)]
            y_train_true = [np.argmax(i) for i in y_train]
            score_train = model.evaluate(x_train, y_train)
            train_loss = score_train[0]
            train_accuracy = score_train[1]
            CM_train = cm(y_train_true, y_train_pred)

            y_test_pred = [np.argmax(i) for i in model.predict(x_test)]
            y_test_true = [np.argmax(i) for i in y_test]
            score_test = model.evaluate(x_test, y_test)
            test_loss = score_test[0]
            test_accuracy = score_test[1]
            CM_test = cm(y_test_true, y_test_pred)

            properties = (i, lr, bt_sz, train_accuracy, train_loss, test_accuracy, test_loss, CM_train, CM_test)
            all_models1.append(properties)
```

```python
# Two-layered NN:

all_models2 = []
for i in number_of_neurons:
    for j in number_of_neurons:
        for lr in learning_rate:
            for bt_sz in batch_size:
                model = Sequential()
                model.add(Dense(i, input_shape=(input_dimension,), activation='sigmoid'))
                model.add(Dense(j, activation='sigmoid'))
                model.add(Dense(num_classes))
#               model.summary()
                model.compile(optimizer=SGD(lr=lr), loss='mse', metrics=['accuracy'])
                model.fit(x_train, y_train, batch_size=bt_sz, epochs=repeat)

                y_train_pred = [np.argmax(i) for i in model.predict(x_train)]
                y_train_true = [np.argmax(i) for i in y_train]
                score_train = model.evaluate(x_train, y_train)
                train_loss = score_train[0]
                train_accuracy = score_train[1]
                CM_train = cm(y_train_true, y_train_pred)

                y_test_pred = [np.argmax(i) for i in model.predict(x_test)]
                y_test_true = [np.argmax(i) for i in y_test]
                score_test = model.evaluate(x_test, y_test)
                test_loss = score_test[0]
                test_accuracy = score_test[1]
                CM_test = cm(y_test_true, y_test_pred)

                properties = (i, j, lr, bt_sz, train_accuracy, train_loss, test_accuracy, test_loss, CM_train, CM_test)
                all_models2.append(properties)
```

```python
train_accuracies1 = [all_models1[i][4] for i in range(len(all_models1))]
test_accuracies1 = [all_models1[i][6] for i in range(len(all_models1))]
print(max(train_accuracies1))
print(max(test_accuracies1))
max_train_acc1 = np.argmax(train_accuracies1)
max_test_acc1 = np.argmax(test_accuracies1)
# print(all_models1[max_train_acc1])
# print(all_models1[max_test_acc1])



train_accuracies2 = [all_models2[i][5] for i in range(len(all_models2))]
test_accuracies2 = [all_models2[i][7] for i in range(len(all_models2))]
print(max(train_accuracies2))
print(max(test_accuracies2))
max_train_acc2 = np.argmax(train_accuracies2)
max_test_acc2 = np.argmax(test_accuracies2)
# print(all_models2[max_train_acc2])
# print(all_models2[max_test_acc2])
```
